```python
score_data = score_data[(score_data['Age'] < 100) & (score_data['Age'] > 0)]
score_data = score_data[(score_data['Delay_from_due_date'] > 0) & (score_data['Num_of_Delayed_Payment'] > 0) & (score_data['Total_EMI_per_month'] > 0)]

```

```python
score_data.Age=np.log(score_data.Age)
score_data.Annual_Income=np.log(score_data.Annual_Income)
score_data.Outstanding_Debt=np.log(score_data.Outstanding_Debt)
score_data.Monthly_Balance=np.log(score_data.Monthly_Balance)
score_data.Delay_from_due_date=np.log(score_data.Delay_from_due_date)
score_data.Credit_History_Age=np.log(score_data.Credit_History_Age)
score_data.Num_of_Delayed_Payment=np.log(score_data.Num_of_Delayed_Payment)
score_data.Total_EMI_per_month=np.log(score_data.Total_EMI_per_month)
```

```python
score_data=score_data.dropna()
score_data.isnull().sum()
```

```python
X=score_data.drop('Credit_Score',axis=1)
y=score_data['Credit_Score']
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y, test_size=0.2, random_state=42)

```

```python
# Create the k-NN classifier
k = 3  # You can adjust the number of neighbors
knn = KNeighborsClassifier(n_neighbors=k)
# Fit the model
knn.fit(X_train, y_train)
# Make predictions
y_pred = knn.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)
y_pred = dt_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
